```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sin(theta):
    return np.sin(theta * np.pi / 180)

def cos(theta):
    return np.cos(theta * np.pi / 180)

def rotation(theta):
    return np.array([[cos(theta), -sin(theta)],
                     [sin(theta), cos(theta)]])

x = np.linspace(0, 5, 100)
y = np.linspace(5, 0, 100)

plt.plot(range(len(x)), x)
plt.plot(range(len(x)), y)

index = 0
for i in range(len(x)):
    d = (x[i]-y[i])**2
    if d < 1e-3*8:
        index = i

inter = np.array([[index],[x[index]]])

plt.scatter(inter[0,0], inter[1,0])

# after = np.dot( rotation(30), inter )
# plt.scatter(after[0,0],after[1,0])
```
