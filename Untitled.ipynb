```python
def F(t):
    return np.array([[1, t],
                     [0, 1]])

def B(t):
    return np.array([[(t**2)/2],
                     [    t   ]])

def U(a):
    return a

def inverse(array):
    return np.linalg.inv(array)

def dot(p, *args):
    for arg in args:
        p = np.dot(p, arg)
    return p

def initCovarianceMatrix(array):
    std_x = array[:, 0].std()
    std_y = array[:, 1].std()
    var_x = std_x**2
    var_y = std_y**2
    covar = std_x * std_y
    return np.array([[var_x, covar],
                     [covar, var_y]])

def initStateMatrix(x, y):
    return np.array([[x],
                     [y],
                     [np.sqrt(x**2 + y**2)/0.01]])

def collectionXYdata(array, x_data=[], y_data=[]):
    h, w = array.shape
    if (h * w) == 4:
        x_data.append(array[0,0])
        y_data.append(array[1,1])
    else:
        x_data.append(array[0,0])
        y_data.append(array[1,0])
    return x_data, y_data
```
